```python
import numpy as np
import matplotlib.pyplot as plt

normal1 = np.random.normal(loc=10, size=1000)
plt.hist(normal1)

X =  np.array([148, 154, 158, 160, 161, 162, 166, 170, 182, 195, 236])

from scipy import stats

res = stats.shapiro(X)
print('res1', res)

res2 = stats.shapiro(normal1)
print('res2', res2)
```
